```python
# --- Baseline packages ---
import pandas as pd
pd.set_option('display.max_columns', None)

# --- Utility packages ---
import os
import sys

# --- Helper functions ---
sys.path.append(os.path.abspath('../src'))
import ecg_dataset_utils
```

# **A.** Preliminaries

```python
# Identify relevant folder and contents.
raw_data_dir = os.path.join("..", "raw-data", "ecg-data-denoised")
filename_list = sorted(os.listdir(raw_data_dir))

# Define key variables.
patient_idx = 0
sampling_rate = 500

# Get data.
patient_filepath = os.path.join(raw_data_dir, filename_list[patient_idx])
rawDataDF = pd.read_csv(patient_filepath, header=None)
ecgDataDF = rawDataDF.T / 1000
ecgDataMat = ecgDataDF.values
```

# **B.** Layout 1

```python
ecg_dataset_utils.ecg_plot(
    ecgDataMat,
    sample_rate = sampling_rate,
    chart_title = None,
    lead_index = ['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6'],
    lead_order = None,
    num_columns = 4,
    row_height = 6,
    short_strip_length = 2.5,
    rhythm_strip_label = 'II',
    show_grid = True,
    style = None,
    line_width = 0.5
)
```
